```python
def longest_substring(s: str) -> int:
    # Initialize a set to store the characters in the current window
    char_set = set()
    # Initialize two pointers to keep track of the window
    i, j = 0, 0
    # Initialize a variable to store the maximum length of the substring
    max_len = 0
    
    while j < len(s):
        # If the current character is not in the set, add it and move the right pointer
        if s[j] not in char_set:
            char_set.add(s[j])
            j += 1
            max_len = max(max_len, j - i)
        # If the current character is in the set, remove the leftmost character and move the left pointer
        else:
            char_set.remove(s[i])
            i += 1
    return max_len

print(longest_substring("abacadaabadc"))
```

```python
import re

def myAtoi(s):
    # Step 1: Read in and ignore any leading whitespace
    s = s.lstrip()
    
    # Step 2: Check if the next character is '-' or '+'
    sign = 1
    if s and (s[0] == '-' or s[0] == '+'):
        sign = -1 if s[0] == '-' else 1
        s = s[1:]
    
    # Step 3: Read in next the characters until the next non-digit character or the end of the input is reached
    digits = re.findall(r'^\d+', s)
    if not digits:
        return 0
    else:
        digits = digits[0]
    
    # Step 4: Convert these digits into an integer
    num = int(digits)
    
    # Step 5: If the integer is out of the 32-bit signed integer range [-2^31, 2^31 - 1], then clamp the integer
    INT_MIN = -2**31
    INT_MAX = 2**31 - 1
    if num * sign < INT_MIN:
        return INT_MIN
    elif num * sign > INT_MAX:
        return INT_MAX
    else:
        return num * sign

print(myAtoi ("42 with words"))

print(myAtoi ("-42"))
```

```python
def min_window(s: str, t: str) -> str:
    t_count = {}
    for c in t:
        if c not in t_count:
            t_count[c] = 1
        else:
            t_count[c] += 1
    left, right = 0, 0
    curr_count = {}
    required = len(t_count)
    formed = 0
    ans = (float('inf'), 0, 0)
    while right < len(s):
        c = s[right]
        if c in t_count:
            if c not in curr_count:
                curr_count[c] = 1
            else:
                curr_count[c] += 1
            if curr_count[c] == t_count[c]:
                formed += 1
        while formed == required:
            if right - left + 1 < ans[0]:
                ans = (right - left + 1, left, right)
            c = s[left]
            if c in t_count:
                curr_count[c] -= 1
                if curr_count[c] < t_count[c]:
                    formed -= 1
            left += 1
        right += 1
    return "" if ans[0] == float('inf') else s[ans[1]:ans[2] + 1]

print(min_window("ADOBECODEBANC","ABC"))
```

```python
def firstUniqChar(s: str) -> int:
    # Create a dictionary to store the count of each character in the string
    char_count = {}
    for char in s:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    # Iterate through the string and return the first character that has a count of 1
    for i, char in enumerate(s):
        if char_count[char] == 1:
            return i
    # If no unique characters are found, return -1
    return -1

print(firstUniqChar("leetcode"))
print(firstUniqChar("loveleetcode"))
```

```python
def longestSubstring(s, k):
    for c in set(s):
        if s.count(c) < k:
            return max(longestSubstring(t, k) for t in s.split(c))
    return len(s)

print(longestSubstring("aaabb", 3))
print(longestSubstring("ababbc", 2))
```
